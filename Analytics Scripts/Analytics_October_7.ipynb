```python
# create a DataFrame
import scipy.stats as ss
import random
def distance_map(distance):
  if 0<=distance<50:
    return "0-50"
  elif 50<=distance<100:
    return "50-100"
  elif 100<=distance<150:
    return "100-150"
  elif 150<=distance<200:
    return "150-200"
  elif 200<=distance<250:
    return "200-250"
  elif 250<=distance<300:
    return "250-300"
  else:
    return "300-350"

distances = ["0-50","50-100","100-150","150-200","200-250","250-300"]
values = [12,15,35,45,24,15]
output = []
for value in values:
  for distance in distances:
    for level in range(1,5):
      output.append([level,distance,value+((5-level)*3)+random.randint(1,20)])


df = pd.DataFrame(output,columns = ['Level','Distance','Count'])
sns.barplot(data=df, x='Distance', y='Count', hue='Level').set_title("Distances covered per level")
```

```python
import pandas as pd
import numpy as np
import altair as alt

index_level = ["Level 1","Level 2","Level 3","Level 4"]
columns = ["First","Last","Middle"]
```

```python
df1=pd.DataFrame(10*np.random.rand(4,3),index=index_level,columns=columns)
df2=pd.DataFrame(10*np.random.rand(4,3),index=index_level,columns=columns)


def prep_df(df, name):
    df = df.stack().reset_index()
    df.columns = ['c1', 'c2', 'values']
    df['DF'] = name
    return df

df1 = prep_df(df1, 'Hits')
df2 = prep_df(df2, 'Misses')


print(df1)
df = pd.concat([df1, df2])
```

```python
alt.Chart(df).mark_bar().encode(

    # tell Altair which field to group columns on
    x=alt.X('c2:N', title=None),

    # tell Altair which field to use as Y values and how to calculate
    y=alt.Y('sum(values):Q',
        axis=alt.Axis(
            grid=False,
            title=None)),

    # tell Altair which field to use to use as the set of columns to be  represented in each group
    column=alt.Column('c1:N', title=None),

    # tell Altair which field to use for color segmentation 
    color=alt.Color('DF:N',
            scale=alt.Scale(
                # make it look pretty with an enjoyable color pallet
                range=['#96ceb4', '#ffcc5c','#ff6f69'],
            ),
        ))\
    .configure_view(
        # remove grid lines around column clusters
        strokeOpacity=0    
    )
```
